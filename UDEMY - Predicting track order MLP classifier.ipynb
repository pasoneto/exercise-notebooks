```python
#import sklearn
import os
import pandas as pd
import numpy as np
os.chdir('C:/Users/Lenovo/Documents/GitHub/music_order/sojazz')
base = pd.read_csv('final_final.csv')
#base.dropna(axis='rows', inplace = True)
#classe = base['posicao_track']
#base.drop(['track_number'], axis=1, inplace=True)
#base.drop(['posicao_track'], axis=1, inplace=True)
#base[['album_id', 'energy', 'posicao_track', 'track_number']]

```

```python
base = base[[#Input
    
             'danceability','energy', 'tempo', 'loudness', 'valence', 
             
             'energy_categorical', 'tempo_categorical', 'valence_categorical', 
            
          
             'proximo_energy', 'proximo_tempo', 'proximo_loudness', 'proximo_danceability',
             
             'loudness_max', 'tempo_max', 
             
             #Output
             
             'energy_next', 'tempo_next', 'loudness_next', 'valence_next',
             
             'proximo_tempo_max', 'proximo_loudness_max',
    
             'posicao_track', 'track_number']]

base
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

codigo = LabelEncoder()
classe = codigo.fit_transform(classe)
base['energy_categorical'] = codigo.fit_transform(base['energy_categorical'])
base['loudness_categorical'] = codigo.fit_transform(base['loudness_categorical'])
base['tempo_categorical'] = codigo.fit_transform(base['tempo_categorical'])
base['valence_categorical'] = codigo.fit_transform(base['valence_categorical']) 

scaler = StandardScaler()
base = scaler.fit_transform(base)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(base, classe)
```

```python
from sklearn.neural_network import MLPClassifier
modelo = MLPClassifier(hidden_layer_sizes=(100, 100), 
                       max_iter=2000, 
                       activation = 'logistic',
                       solver='lbfgs', 
                       random_state=1)
modelo.fit(x_train, y_train)
```

```python
########## MAKE PREDICTIONS ##################
previsoes = modelo.predict(x_test)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(modelo, base, classe, cv=10)
cv_mean = np.mean(scores)
cv_std = np.std(scores)
cv_mean
```
