```python
import librosa
import matplotlib.pyplot as plt
import pandas as pd
from librosa import display
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", 1000)
```

```python
#Baixando audio
path = 'C:/Users/Lenovo/Desktop'
audio_1 = '/500.wav'
audio_2 = '/300.wav'
audio_3 = '/500_600_700.wav'

s_500 , sr = librosa.load(path+audio_1)
s_300, sr = librosa.load(path+audio_2)
complexo, sr = librosa.load(path+audio_3)
```

```python
#Olhando para um per√≠odo do sinal

n0 = 0 #chosse lower end
n1 = 45 #choose upper end in ms
plt.figure(figsize=(14, 5))
plt.plot(complexo[n0:n1])
plt.grid()
```

```python
## ALGORITMO 1 ##################
#Documentacao
#https://librosa.github.io/librosa/0.7.2/generated/librosa.core.piptrack.html

pitches_simples, magnitudes1 = librosa.piptrack(y=s_300, sr=sr)
pitches_complexo, magnitudes2 = librosa.piptrack(y=complexo, sr=sr)

plt.figure()
librosa.display.specshow(magnitudes1, y_axis='cqt_hz')
```

```python
############# ALGORITMO 2 ##################
#Documentacao

chroma_stft = librosa.feature.chroma_stft(y=complexo, sr=sr, n_chroma=12, n_fft=4096)
chroma_cq = librosa.feature.chroma_cqt(y=complexo, sr=sr)

plt.figure()
librosa.display.specshow(chroma_cq, y_axis='chroma')
```
