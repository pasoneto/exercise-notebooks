```python
import requests
import pandas as pd
```

```python
#Para buscar uma cidade ou local, chame a função abaixo com o nome da cidade em string.
#Para nome composto, use "%20" como espaco
def vendo_estacao(nome_cidade):
    return requests.get(f"https://api.meteostat.net/v1/stations/search?q={nome_cidade}&key=fNHHoPG4")

#Exemplo com belo horizonte
estacao = vendo_estacao("belo%20horizonte")

#Essa função retorna uma resposta crua, que você não consegue usar sem transformar (faco isso abaixo)
```

```python
#Estações proximas de Belo Horizonte
estacoes = estacao.json() # Estações proximas de Belo Horizonte
#Transformando para data frame
estacoes = pd.DataFrame(estacoes['data'])

#Visualizando
estacoes

```

```python
#Agora você precisa pegar o id do local que te interessa
#Fazemos isso pelo pandas
aeroporto_bh = estacoes['id'][2]
```

```python
#Agora definimos uma funcao para buscar os dados climáticos do aeroporto de BH
#Para um determinado range de datas (primeiro ano, depois mes, separado por "-")

#Depois verifique qual é o timezone que você quer
#A função abaixo retorna os horários da europa, mas tem como mudar
def pegando_dados(local, inicio: str, fim: str):
    return requests.get(f"https://api.meteostat.net/v1/history/hourly?station={local}&start={inicio}&end={fim}&time_zone=Europe/London&time_format=Y-m-d%20H:i&key=fNHHoPG4").json()
```

```python
inicio = "2020-01-01"
fim = "2020-01-02"
pegando_dados(aeroporto_bh, inicio, fim)
```
