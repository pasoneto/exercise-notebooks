```python
#Cleaning up strings
def tira(vetor):
    if type(vetor) == int:
        return vetor
    if type(vetor) == float:
        return vetor
    else:
        return vetor.strip("/'/[]/")

    
#Getting information about strings
def barra(vetor):
    if type(vetor) == int:
        return vetor
    if type(vetor) == float:
        return vetor
    if vetor.find('/') == -1:
        return vetor
    if vetor.find('/') != -1:
        posicao = vetor.find('/')
        return vetor[0:posicao]

def traco(vetor):
    if type(vetor) == int:
        return vetor
    if type(vetor) == float:
        return vetor
    if vetor.find('-') == -1:
        return vetor
    if vetor.find('-') != -1:
        posicao = vetor.find('-')
        return vetor[0:posicao]

def virgula(vetor):
    if type(vetor) == int:
        return vetor
    if type(vetor) == float:
        return vetor
    if vetor.find(',') == -1:
        return vetor
    if vetor.find(',') != -1:
        posicao = vetor.find(',')
        return vetor[0:posicao]

def bpm(vetor):
     if type(vetor) == int:
        return vetor
     if type(vetor) == float:
        return vetor
     if vetor.find('b') == -1:
        return vetor
     if vetor.find('b') != -1:
        posicao = vetor.find('b')
        return vetor[0:posicao]
    
def bpm_maiusculo(vetor):
     if type(vetor) == int:
        return vetor
     if type(vetor) == float:
        return vetor
     if vetor.find('B') == -1:
        return vetor
     if vetor.find('B') != -1:
        posicao = vetor.find('B')
        return vetor[0:posicao]
```
