```python
oi = df[~df.ips.isna()]
```

```python
oi = oi[oi.ips != "noIP"]
```

```python
def url(ip):
    return f'https://api.ipgeolocation.io/ipgeo?apiKey=3f18a34148554041920e68b6cab024fe&ip={ip}' 
    
```

```python
test = url("189.54.146.48")
```

```python
r = requests.get(test)
```

```python
json.loads(r.content)['country_name']
```

```python
def cenoura(ip):
    r = requests.get(url(ip))
    j = json.loads(r.content)
    return j.get("country_name", "")
```

```python
cenoura("189.54.146.48")
```

```python
oi["pais"]=oi.ips.apply(cenoura)
```
