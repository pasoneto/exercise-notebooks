```python
#Dividindo o banco de dados em previsores e outcomes (classificadores)
previsores = base.iloc[:, 1:4].values
classe = base.iloc[:, 4].values


#Preenchendo os valores faltantes. Usamos o imputer do sklearn, que é um
#algorítmo que faz o input baseado em algum parâmetro dos dados
imputer = Imputer(missing_values = 'NaN', strategy = 'mean', axis = 0)
imputer = imputer.fit(previsores[:, 0:3])
previsores[:, 0:3] = imputer.transform(previsores[:, 0:3])
```

```python
#Normalização, padronização e standardization
#Padronização é a observação menos a média do grupo, dividido pelo Standard Deviation
#Abaixo ele está aplicando a formula descrita acima ao banco de dados.
scaler = StandardScaler()
previsores = scaler.fit_transform(previsores)
```
