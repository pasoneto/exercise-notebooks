```python
#Function to get list of data frames
def get_list_df(df):
    final = []
    albums = list(df['album_id'].unique())
    grouped = df.groupby(['album_id'])
    for i in albums:
        final.append(grouped.get_group(i))
    return final
```

```python
random.shuffle(base)
#base = [i for i in base]
feature_cols = ['danceability', 'tempo_overall', 'loudness_overall', 'valence', 'energy', 'energy_diff', 'valence_diff', 'loudness_diff', 'int_energy_valence_loudness', 'int_energy_valence', 'int_valence_loudness', 'int_energy_loudness']
# 'tempo_categorical', 'valence_categorical', 'energy_categorical', 'loudness_categorical',
#Defining sample with track_number as input and feature as output
def sample_making(lista_de_albums):
    entradas = []
    saidas = []
    for k in lista_de_albums:
        entradas.append(k[feature_cols].sample(frac=1))
        saidas.append(k[["energy", "loudness_overall"]])
    return entradas, saidas
```

```python
#Treino e teste
#random.shuffle(base)
entrada_treino = base[0:2275] 
entrada_teste = base[2276:len(base)]

entrada_treino, saida_treino = sample_making(entrada_treino)
entrada_teste, saida_teste = sample_making(entrada_teste)
```

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```

```python
energy = pd.DataFrame({"energy_prev" : energy_prev,
                       "energy_real": energy_real})
loudness = pd.DataFrame({"loudness_prev" : loudness_prev,
                       "loudness_real": loudness_real})
# valence = pd.DataFrame({"valence_prev" : valence_prev,
#                        "valence_real": valence_real})
```
