```python
feature = ['barkbands_spread','melbands_spread','hfc','barkbands_kurtosis','spectral_energyband_middle_low','melbands_crest','spectral_kurtosis','spectral_rms','zerocrossingrate','silence_rate_60dB','erbbands_kurtosis','spectral_strongpeak','spectral_energy','spectral_rolloff','spectral_centroid','pitch_salience','spectral_energyband_middle_high','melbands_skewness','spectral_spread','dissonance','spectral_skewness','spectral_flux','erbbands_flatness_db','spectral_energyband_high','spectral_entropy','silence_rate_20dB','melbands_flatness_db']
estatisticas = ['dmean2', 'median', 'min', 'dvar2', 'dvar', 'dmean', 'max', 'var', 'mean'] 
feature_tonal = ['chords_number_rate', 'key_scale', 'key_key', 'chords_changes_rate', 'key_strength']

def data(path):
    try:
        root = json.load(open(path))
        part_1 = {x : root['rhythm'][x] for x in ['bpm', 'danceability']}
        part_2 = {f'{f}_{s}' : root['lowlevel'][f][s] for f, s in product(feature, estatisticas)}
        part_3 = {x: root['tonal'][x] for x in feature_tonal}
        part_5 = {x: root['lowlevel'][x] for x in ['average_loudness', 'dynamic_complexity'] }
        return {**part_1, **part_2, **part_3, **part_5, **root['metadata']['tags']}
    except:
        return {}    

def process_folder(path, output_file):
    rows_dicts = map(data, path.iterdir())
    df = pd.DataFrame.from_records(rows_dicts)
    df.to_csv(output_file, index=False)
    print(f'folder {path.name} done')


folders_dir = Path('C:/Users/Lenovo/Desktop/world/Ciência/Projetos de pesquisa/Estudo Computacional/acousticbrainz-mediaeval-train')
caminho = 'C:/Users/Lenovo/Desktop/world/Ciência/Projetos de pesquisa/Estudo Computacional/output'
for folder in folders_dir.iterdir():
    process_folder(folder, f'{caminho}/folder_{folder.name}.csv') 
```
